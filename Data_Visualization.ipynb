```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
#loading the iris dataset from sklearn liabrary
iris=snr.load_dataset('iris')
```

visualizing the dataset

```python
#line graph
ax=iris.plot(figsize=(15,8),title='visualization on iris dataset')
ax.set_xlabel('x_axis') #rows
ax.set_ylabel('y_axis') #columns
```

```python
#dropping the categorical variable 'species'
data=iris.drop(['species'],axis=1) 
```
