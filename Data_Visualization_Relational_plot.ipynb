```python
import numpy as np
import pandas as pd
import seaborn as snr
import matplotlib.pyplot as plt
```

Visualizing the Dataset using **Relational plot**

```python
#Relational Plot is used to compare more than 2 or 3 parameter
```

```python
data=pd.read_csv('/content/tips.csv')
```

```python
data.head()
```

```python
# will compare sex and tip by using scatter plot
data.plot.scatter(x='tip',y='sex')
```

```python
#will compare total bill and sex using relplot
snr.relplot(x='total_bill',y='sex', data=data)
```

```python
snr.relplot(x='tip',y='sex', data=data)
```

```python
#comparing 3 parameter
snr.relplot(x='total_bill',y='tip', hue='smoker',data=data)
```

```python
#compare 4 parameter
snr.relplot(x='total_bill',y='tip', hue='sex',style='time', data=data)
```

```python
#compare 5 parameter
snr.relplot(x='total_bill',y='tip', hue='smoker',style='time', col='sex',data=data)
```

```python
#The graph below compares the times when customers have visited the restaurant the most.
snr.relplot(x='total_bill',y='tip', hue='smoker',style='sex', col='time',data=data)
```

```python
#comparing 5 parameter
snr.relplot(x='total_bill',y='tip', hue='smoker',style='sex', size='size',data=data)
```
